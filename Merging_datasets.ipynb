```python
df_2023d = pd.read_csv('./StormEvents_details_2023.csv')
df_2023d.head()
```

```python
df_2023d.info()
```

```python
df_2023f = pd.read_csv('./StormEvents_fatalities_2023.csv')
df_2023f.info()
```

```python
df_2023l = pd.read_csv('./StormEvents_locations_2023.csv')
df_2023l.info()
```

```python
first_join = pd.merge(df_2023d, df_2023f, on="EVENT_ID", how='left')
first_join.head()
```

```python
second_join = pd.merge(first_join, df_2023l, on='EVENT_ID', how='left')
second_join.head()
```

```python
final_df = pd.concat([second_merge, second_join])
final_df.info()
```

```python
final_df.to_csv("stormdata.csv", index=False)
```
